First we'll do our **`replace()`** function, which should take in 3 strings, one to search in, the second which is the desired string we wish to replace, and the third is the string that the previous should be replaced by

```python
def replace(string, sub, new):
    wordsInString = string.split()
    # Loops through all words in word list
    i = 0
    while i < len(wordsInString):
        if wordsInString[i] == sub: #If word selected in loop is the word we wish to substitute
            wordsInString[i] = new # Replaces word for new word
        i+=1 # Important so we actually move on to the next word

    #I had to look up a function to add my strings inside of my list back into a sentence. -> https://www.w3schools.com/python/ref_string_join.asp
    newstr = " ".join(wordsInString)
    print(newstr)
        
```

**Now we'll do our `count()` function**, which should take in 2 strings, one to search in, and one string to count the number of occurences of in the previous string.

```python
def countWord(sentence, sub):
    count = 0
    wordsInString = sentence.split()
    # Found a better way to loop with index -> https://www.w3schools.com/python/python_lists_loop.asp.
    # For loop
    for i in range(len(wordsInString)):
        if wordsInString[i] == sub:
            count += 1
    print("The word " + sub + " appears " + str(count) + " times in sentence '" + sentence + "'.")

countWord("I really really really really really really really like coding", "really")
```

For example, is_balanced("(8x + (46^3)) = 24") returns True, but is_balanced(")()(") returns False. If you want an added challenge, add the functionality for braces {} and brackets [].

```python
def longest(sentence):
    # Find longest word
    wordsInString = sentence.split()
    biggestWord = ""
    # For loop
    for i in range(len(wordsInString)):
        # if our word is longer that our current biggest word...
        if len(wordsInString[i]) > len(biggestWord):
            biggestWord = wordsInString[i] # Then it's our new biggest word
    
    #Find where the word starts and ends in our sentence
    startOfWord = sentence.find(biggestWord)
    endOfWord = startOfWord + len(biggestWord)

    print("The biggest word's tuple is [" + str(startOfWord) + "," + str(endOfWord) + "]")

# TESTS :)
longest("Very cool sentence with this very long word called deoxyribonucleic ") #Ans should be [51,67]
longest("Have I mentioned I love my snacks? ") #Ans should be [7,16]

def is_balanced(sentence):
    # Simplest solution I can think of is having a balance checker variable that increases or decreases with each closing or starting parentheses/bracket/curly bracket
    parenthesesBalance = 0
    curlyBalance = 0
    bracketBalance = 0
    for char in sentence:
        if char == "(":
            parenthesesBalance += 1
        elif char == ")":
            parenthesesBalance -= 1
        elif char == "{":
            curlyBalance += 1
        elif char == "}":
            curlyBalance -= 1
        elif char == "[":
            bracketBalance += 1
        elif char == "]":
            bracketBalance -= 1
        
        if bracketBalance < 0 or curlyBalance < 0 or parenthesesBalance < 0:
            return(False)

    if bracketBalance == 0 and curlyBalance == 0 and parenthesesBalance == 0:
        return(True)
    else:
        return(False)

# TESTS :)
print(is_balanced("(8x + (46^3)) = 24"))  # True
print(is_balanced(")()("))                 # False
print(is_balanced("((()))"))               # True
print(is_balanced("(()"))                  # False
print(is_balanced("())"))                  # False
```
