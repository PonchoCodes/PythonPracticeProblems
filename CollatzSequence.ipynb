## **<center> Problem 3** </center>
Write a function that takes in a positive integer n
, then outputs the Collatz sequence for that number:
If the number n=1
, you’re done.
If the number n
 is even, divide it by 2 to get the next number in the sequence: n→n2
.
If the number n
 is an odd number bigger than 1, multiply by 3 and add 1 to get the next number in the sequence: n→3⋅n+1
.
Then, repeat the process until you reach 1, and output the list of numbers along the way to 1. For example, for collatz(3), the output should be [3, 10, 5, 16, 8, 4, 2, 1].

First I made my Collatz Sequence function as observed below

```python
def CollatzSequence(inputN):
    # If number is already 1 simply output 1
    if inputN == 1:
        print("Your number is already 1")
    
    # If number isn't 1
    elif inputN != 1:
        collatz = []
        n = inputN
        while n != 1: # Loops until we get to 1
            if n % 2 != 0: #If number is odd
                n = (n * 3) + 1 # Multiply by 3 add 1
                collatz.append(n) # Add number to Collatz Sequence
            else: #If number is even
                n = n/2 # Divide number by 2
                collatz.append(n) # Add number to Collatz Sequence

        print("Your Collatz Sequence is:" + str(collatz))
        
```

And, then I simply tested it to make sure it worked!

```python
# Should output 1   
CollatzSequence(1)

# Anything else should output a sequence that eventually gets to 1
CollatzSequence(2) #Even number test
CollatzSequence(3) #Odd number test
CollatzSequence(2231) #Big Number test
CollatzSequence(2313023213) #Huge Number test
CollatzSequence(2221394387159817571589356754569021491240124831) #Massive Number test
```

Sure enough every Collatz Sequence from our test results gets us to a 1!
